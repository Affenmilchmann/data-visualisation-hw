```python
import pandas as pd
from matplotlib import pyplot as plt
df = pd.read_csv('Pearson.tsv', sep='\t')

total_father = df['Father']
total_son = df['Son']

plt.scatter(total_father, total_son)
```

```python
df1 = df[(65 - 0.5 < df['Father']) & (df['Father'] < 65 + 0.5)]
df2 = df[(67.5 - 0.5 < df['Father']) & (df['Father'] < 67.5 + 0.5)]
df3 = df[(70 - 0.5 < df['Father']) & (df['Father'] < 70 + 0.5)]

df1['Son'].to_csv('65.csv', index=False)
df2['Son'].to_csv('67_5.csv', index=False)
df3['Son'].to_csv('70.csv', index=False)
```
